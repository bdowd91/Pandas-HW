```python
import pandas as pd
```

```python
pcs_school_data_to_load = "Pandas-HW/PyCitySchools/PCS_Summary.csv"
```

```python
pcs_student_data_to_load = "Pandas-HW/PyCitySchools/PCS_Students.csv"
```

```python
school_data = pd.read_csv(pcs_school_data_to_load)
student_data = pd.read_csv(pcs_student_data_to_load)
```

```python
copy_school_data = school_data.copy()
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
total_schools = school_data["school_name"].count()
print(total_schools)
```

```python
total_students = student_data["school_name"].count()
print(total_students)
```

```python
total_budget = school_data["budget"].sum()
```

```python
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()
```

```python
math_pass = student_data.loc[(student_data["math_score"] >= 70)]
count_pass_math = math_pass["math_score"].count()
print(count_pass_math)

per_math_pass = (count_pass_math/total_students)*100
print(per_math_pass)
```

```python
read_pass = student_data.loc[(student_data["reading_score"]>= 70)]
count_pass_read = read_pass["reading_score"].count()
print(count_pass_read)

per_read_pass = (count_pass_read/total_students)*100
print(per_read_pass)
```

```python
overall_pass = (per_math_pass + per_read_pass)/2
print(overall_pass)
```

```python
district_summary = {"Total Schools" : total_schools, "Total Students" : total_students, "Total Budget": total_budget, "Average Math Score" : avg_math_score, "Average Reading Score" : avg_reading_score, "% Passing Math" : per_math_pass, "% Passing Reading" : per_read_pass, "% Overall Passing" : overall_pass}
```

```python
district_summary_df = pd.DataFrame([district_summary])
```

```python
district_summary_df = district_summary_df[["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
```

```python
school_data_2 = school_data.copy()
```

```python
school_data_2['Per Student Budget'] = school_data_2['budget']/school_data_2['size']
```

```python
avg_math_read_tbl = student_data.groupby(['school_name'])['reading_score', 'math_score'].mean().reset_index()
```

```python
avg_math_read_tbl
```

```python
school_data_2 = school_data_2.merge(avg_math_read_tbl, on='school_name', how="outer")
school_data_2
```

```python
summary_passing_reading = student_data[student_data['reading_score']>=70]
summary_passing_math = student_data[student_data['math_score']>=70]
```

```python
pass_read_count_sum = summary_passing_reading.groupby(["school_name"])['reading_score'].count().reset_index()
```

```python
pass_math_count_sum = summary_passing_math.groupby(["school_name"])['math_score'].count().reset_index()
```

```python
pass_count = pass_math_count_sum.merge(pass_read_count_sum, on="school_name", how='inner')
```

```python
school_data_2 = school_data_2.merge(pass_count, on="school_name", how='outer')
```

```python
school_data_2['% Passing Math'] = (school_data_2['math_score']/school_data_2['size'])*100
```

```python
school_data_2['% Passing Reading'] = (school_data_2['reading_score']/school_data_2['size'])*100
```

```python
school_data_2['% Overall Passing'] = (school_data_2['% Passing Math'] + school_data_2['% Passing Reading'])/2
```

```python
school_data_2
```
